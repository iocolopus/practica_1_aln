```python
import numpy as np
import matplotlib.pyplot as plt

# Configuramos los parámetros iniciales
np.random.seed(0)
t = np.arange(0, 100, 1)  # Instantes de tiempo (100 pasos)

# Simulación de 4 poblaciones en cada instante de tiempo
# La suma de las poblaciones en cada instante debe ser constante o limitada (p.ej. 1000 individuos)
n_poblaciones = 4
poblaciones = np.random.dirichlet(np.ones(n_poblaciones), size=len(t)) * 1000

# Definimos los colores para cada población
colores = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

# Creamos el gráfico apilado (stacked area plot)
plt.figure(figsize=(10, 6))
plt.stackplot(t, poblaciones.T, labels=[f'Población {i+1}' for i in range(n_poblaciones)], colors=colores)

# Configuramos el gráfico
plt.title('Evolución de las poblaciones en un algoritmo evolutivo')
plt.xlabel('Tiempo (t)')
plt.ylabel('Número de individuos')
plt.legend(loc='upper right')
plt.show()
```

```python
matriz = [[10, 40, 90], [20, 40, 10], [70, 20, 0]]

plt.figure(figsize=(10, 6))

plt.stackplot([i + 1 for i in range(len((matriz)))], matriz, colors=["#1e81b0", "#e28743", "#abdbe3"])
plt.title('Evolucion')
plt.xlabel('Tiempo (t)')
plt.ylabel("Poblacion")
plt.legend(["pene1", "pene2", "pene3"], loc='upper right')
plt.show()
```
